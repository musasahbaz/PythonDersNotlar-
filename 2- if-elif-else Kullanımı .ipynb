#### if Kullanımı 

```python
a = 5
b = 10

if a > b :
    print("a sayısı b sayısından büyüktür")
if b > a :
    print("b sayısı a sayısından büyüktür")
```

```python
#Program çalıştığında kullanıcıdan bir sayı istesin ve bu sayının pozitif-negatif durumunu değerlendirsin

sayi = int(input("Bir sayı giriniz : "))    # Kullanıcıdan sayıyı input komutu ile alıyoruz
if sayi > 0 :                               # Eğer sayı 0'dan büyükse ekrana "Pozitif" yazsın
    print("Girilen sayı pozitiftir.")
if sayi < 0 : 
    print("Girilen sayı negatiftir")
if sayi == 0 : 
    print("Girilen sayı sıfırdır.")
```

##### Yukarıdaki örnekte görüldüğü gibi bir hata oluştu. Bu hatayı gidermek için elif kullanımını öğrenelim

### elif Kullanımı

```python
#Klavyeden öğrenci puanı girilecek ve bu puanın not olarak kaça denk geldiği bulunacak.

puan = int(input("Ders puanını giriniz: "))

if puan < 50:
    print("Notunuz: 1")

elif puan < 60:
    print("Notunuz: 2")

elif puan < 70:
   print("Notunuz: 3")

elif puan < 85:
   print("Notunuz: 4")

elif puan <= 100:
    print("Notunuz: 5")

elif puan > 100:
    print("Hatalı giriş!")

elif puan < 0:
    print("Hatalı giriş!")
```

### else Kullanımı 

```python
# Klavyeden girilen öğrencinin notu 50 ve üstünde ise "GEÇTİ"; 50'nin altında ise "KALDI" yazsın
```

```python
puan = int(input("Ders puanını giriniz: "))

if puan > 50:            # Girilen not 50 ve üstünde ise
    print("GEÇTİNİZ")
else :                   # Girilen not 50 ve üstünde değilse
    print("KALDINIZ")
```

```python
# Kullanıcıdan parola girilmesi istensin. Eğer girilen parola 1234 ise ekrana "Sisteme Hoşgeldiniz" yazsın.
# Girilen parola 1234 dışında ise ekrana "Yanlış parola!" yazsın

parola = input("Parola: ")
if parola == "1234":
    print("Sisteme Hoşgeldiniz!")
else :
    print("Yanlış parola!")
```

#### if - elif - else kullanımına dair örnekler

```python
#ÖRNEK - 1 : Klavyeden girilen yaşa göre değerlendirme 

yas = int(input("Yaşınızı giriniz : "))
if yas < 13:
    print("Çocuk")
elif yas < 18:
    print("Genç")
else:
    print("Yetişkin")
```

```python
#ÖRNEK - 2 : Klavyeden girilen sıcaklık değerine göre değerlendirme

sicaklik =  int(input("Sıcaklık değeriniz giriniz : "))
if sicaklik < 10:
    print("Soğuk")
elif sicaklik < 25:
    print("Ilık")
else:
    print("Sıcak")
```

```python
#ÖRNEK - 3 : Klavyeden girilen sıcaklık değerine göre suyun halini bulma

sicaklik =  int(input("Suyun sıcaklık değeriniz giriniz : "))
if sicaklik < 0:
    print("KATI")
elif sicaklik < 100:
    print("SIVI")
else:
    print("GAZ")
```

```python
#ÖRNEK - 4 : Klavyeden girilen sayının pozitif-neagatif durumu

sayi =  int(input("Sayı giriniz : "))
if sayi > 0:
    print("Pozitif")
elif sayi < 0:
    print("Negatif")
else:
    print("Sıfır")
```

```python
#ÖRNEK - 5 : Klavyeden girilen boy değerine göre değerlendirme

boy =  int(input("Boyunuzu cm olarak giriniz : "))
if boy < 150:
    print("Kısa")
elif boy < 180:
    print("Orta")
else:
    print("Uzun")
```

```python
#ÖRNEK - 6 : Klavyeden girilen mesajın uzunluğuna göre değerlendirme

mesaj =  input("Mesajınızı giriniz : ")
if len(mesaj) < 5:
    print("Kısa mesaj")
elif len(mesaj) < 15:
    print("Orta mesaj")
else:
    print("Uzun mesaj")
```

```python
#ÖRNEK - 7 : Klavyeden girilen  şifrenin uzunluğuna göre değerlendirme

sifre =  input("Şifrenizi giriniz : ")
if len(sifre) < 4:
    print("Zayıf şifre")
elif len(sifre) < 8:
    print("Orta şifre")
else:
    print("Güçlü şifre")
```

```python
#ÖRNEK - 8 : Klavyeden girilen yaş değerine göre değerlendirme

yas =  int(input("Yaşınızı giriniz : "))
if yas < 7:
    print("Ücretsiz")
elif yas < 25:
    print("Öğrenci")
elif yas < 60:
    print("Tam")
else:
    print("65 yaş indirimli")
```

```python
#ÖRNEK - 9 : Klavyeden girilen stok adedine göre değerlendirme

stok =  int(input("Stok adetini giriniz : "))
if stok == 0:
    print("Stok yok")
elif stok <= 10:
    print("Az stok")
else:
    print("Stok yeterli")
```

```python
#ÖRNEK - 10 : Klavyeden girilen kargo kilo değerine göre değerlendirme

kilo =  int(input(" Kilonuzu giriniz : "))
if kilo < 1:
    print("Hafif paket")
elif kilo < 5:
    print("Orta paket")
else:
    print("Ağır paket")
```

```python
#ÖRNEK - 11 : Klavyeden girilen sayı değerine göre değerlendirme

sayi =  int(input("Sayı giriniz : "))
if sayi % 2 == 1:
    print("Tek")
elif sayi % 2 == 0:
    print("Çift")
```

```python
#ÖRNEK - 12 : Klavyeden girilen ücrete göre değerlendirme

ucret =  int(input("Ücreti  giriniz : "))
if ucret < 30000:
    print("Düşük gelir")
elif ucret < 70000:
    print("Orta gelir")
else:
    print("Yüksek gelir")
```
