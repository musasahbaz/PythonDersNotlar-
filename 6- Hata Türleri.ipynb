## Python Hata Türleri

3) **TabError:** Sekme (Tab) ve boşlukların (Space) aynı kod bloğunda karıştırılması.
4) **Değişken ve İsimlendirme Hataları**
5) **Veri Tipi ve Değer Hataları**
6) **Matematiksel ve Mantıksal Hatalar** (ZeroDivisionError)
