#### 🔎 2. Karakterlere Erişim (Indexing)

```python
kelime = "Python"

print(kelime[0])   # P
print(kelime[3])   # h
```

```python
ad = "Reşat Nuri Güntekin"
print( ad[0] )           #Burada ad stringinin ilk elemanına (0.eleman) erişildi 🪄
print( ad[7] )          # aynı şekilde ad stringinin 7. elemanına  erişildi
print( ad[-1] )          # Stringin son elemanına (-1) erişmek için bu kod kullanılır.
```

```python
ad = "Reşat Nuri Güntekin"
print( ad[27] )          # eğer stringin uzuluğundan daha yüksek bir karakter sırasına erişmek istersek arlık dışı hatası alırız.
```

#### Negatif indeksler:

####
✔ -1 → son karakter

✔ -2 → sondan ikinci karakter

```python
print(kelime[-1])  # n
print(kelime[-2])  # o
```

```python
isim = input("isminiz: ")                                   #Kullanıcıdan ismini girmesini istedik 🪄
for i in range(len(isim)):                                  #For döngüsü ile ismin tüm harflerine sırayla eriştik
    print("isminizin {} . harfi: {} ".format(i+1, isim[i]))  # Burada i sayaç sayısını, isim[i] ise stringdeki denk geldiği elemanı verir.
```

#### ✂ 3. Dilimleme (Slicing)

Belirli aralıktaki karakterleri almak için kullanılır.

```python
kelime = "Python Programlama"

print(kelime[0:6])   # Python        🪄
print(kelime[7:])    # Programlama
print(kelime[:6])    # Python
```

### 🧰 7. Önemli String Metotları

#### 🔹 lower() / upper()
metni küçük/BÜYÜK harfe çevirir.

```python
print("Python".lower())  # python (Tamamını küçük harfe çevirir.) 🪄
```

```python
print("python".upper())  # PYTHON (Tamamını BÜYÜK harfe çevirir.)
```

#### 🔹 capitalize()  /  title()
Başharfini büyük yapar 

```python
print("türkiye büyük millet meclisi".capitalize())  # Sadece ilk harfi BÜYÜK harfe çevirir.
```

#### 🔹  title()
Her sözcüğün baş harfini büyük harf yapar 

```python
print("türkiye büyük millet meclisi".title())  # Her kelimenin ilk harfini BÜYÜK harfe çevirir.
```

#### 🔹  swapcase()
Metinde yer alan BÜYÜK hafleri küçük harfe; küçük harfleri ise BÜYÜK harfe dönüştürür. 

```python
print("Türkiye Büyük Millet Meclisi".swapcase())  # küçük-BÜYÜK harf değişimi yapar.
```

#### 🔹 strip()
Baştaki ve sondaki boşlukları siler.

```python
metin = "  Python  "
print(metin.strip())      #strip() ile herhangi bir parametre yazılmazsa baştaki ve sondaki boşlukları kırpar
```

```python
metin = "kocakonak"
print(metin.strip("k"))       #strip() ile parametre verilirse verilen karakteri başta ve sonda varsa kırpar
```

#### 🔹 lstrip()
Baştaki ve sondaki boşlukları siler.

```python
metin = "kocakonak"
print(metin.lstrip("k"))       #lstrip() ile soldan istenen kırpmanın yapılması sağlanır
```

#### 🔹 rstrip()
Baştaki ve sondaki boşlukları siler.

```python
metin = "kocakonak"
print(metin.rstrip("k"))       #strip() ile sağdan istenen kırpmanın yapılması sağlanır
```

#### 🔹 replace()

```python
print("Merhaba Dünya".replace("Dünya", "Python"))
```

#### 🔹 split()
Metni listeye çevirir.

```python
cumle = "Python çok güzel"
print(cumle.split())
```

#### 🔹 join()
Listeyi stringe çevirir.

```python
kelimeler = ["Python", "çok", "güzel"]
print(" ".join(kelimeler))
```

#### 🔹 count()
Belirtilen karakterin, metin içerisinde kaç adet bulunduğunu söyler

```python
metin = "Türkiye Büyük Millet Meclisi"
print(metin.count("i"))             # Metin içerisinde geçen "i" harflerinin sayısını verir
```

```python
metin = "Türkiye Büyük Millet Meclisi"
print(metin.count("i",3))             # Metin içerisindeki 3.karakterden itibaren geçen "i" harflerinin sayısını verir
```

```python
metin = "Türkiye Büyük Millet Meclisi"
print(metin.count("i",3,17))             # Metin içerisindeki 3.karakter ile 17.karakter arasında geçen "i" harflerinin sayısını verir
```

####  index() 
Sorguladığımız karakterin, karakter dizisi içinde geçtiği ilk konumu verir

```python
metin = "Türkiye Büyük Millet Meclisi"
print(metin.index("i"))             # Metin içerisindeki ilk "i" harfi 4.karakterdedir.
```

#### 🔎 rindex() 
Sorguladığımız karakterin, sağdan karakter dizisi içinde geçtiği ilk konumu verir. Aramaya metnin sonundan itibaren başlar ve başa doğru devam eder.

```python
metin = "Türkiye Büyük Millet Meclisi"
print(metin.rindex("i"))             # Metin içerisindeki sağdan  ilk "i" harfi 27.karakterdedir.
```

#### 🔎 center()
Metni verilen karakter sayısı kadar soldan girinti bırakarak hizalar

```python
metin = "Türkiye Büyük Millet Meclisi "
print(metin.center(100))
```

#### 🔎 isalpha()
Verilen karakter dizisinin alfabetik olup olmadığını söyler.

Alfabetik karakter dizileri, alfabe harflerinden oluşan karakter dizileridir.

Sayısal karakter dizileri, rakamlardan oluşan karakter dizileridir.

Alfanümerik karakter dizileri ise bunun birleşimidir.

```python
metin = "Telefon"
print(metin.isalpha())   #metin değişkeni string değer taşıyor mu?
```

#### 🔎 isdigit()
Verilen karakter dizisinin sayısal olup olmadığını söyler.

```python
metin = "1245"
print(metin.isdigit())   #metin değişkeni sayısal değer taşıyor mu?
```

#### 🔎 isnumeric()
Bu metot bir karakter dizisinin nümerik olup olmadığını denetler. Yani bu metot yardımıyla
bir karakter dizisinin sayı değerli olup olmadığını denetleyebiliriz:

```python
metin = "1245"
print(metin.isnumeric())   #metin değişkeni sayısal değer taşıyor mu?
```

#### 🔎 isalnum()
Verilen karakter dizisinin alfanumerik olup olmadığını söyler.

```python
metin = "abc123"
print(metin.isalnum())
```

#### 🔎 8. in Operatörü

#### 🧱 9. String Değiştirilemezlik (Immutable)
Bu hata verir:

```python
kelime = "Jython"
kelime[0] = "P"   # 0.karakteri "P" yap   🪄
```

```python
# Doğru yöntem:  🪄
kelime = "Jython"
yeni = "P" + kelime[1:]
print(yeni)
```

#### 🎯 10. Formatlama Yöntemleri

#### 🔹 .format() 

```python
isim = "Albert Camus"
sene = 1913
print("Merhaba {}. Sen {} yılında doğdun".format(isim,sene))
```

#### 🔹 f-string (Modern Yöntem)

```python
isim = "Albert Camus"
sene = 1913
print(f"Merhaba {isim}. Sen {sene} yılında doğdun")
```

```python
isim = "Albert Camus"
print(f"Merhaba {isim:^50} nasılsın?" )
```

####  3️⃣ Büyük harf sayısını bulma

```python
metin = input("Metin girin: ")
sayac = 0

for harf in metin:
    if harf.isupper():
        sayac += 1

print("Büyük harf sayısı:", sayac)
```
