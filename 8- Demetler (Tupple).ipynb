## **Demetler**  (Tupple)

Demetlerin ayırt edici özelliği normal parantez işaretleridir. Ayrıca demetler değiştirilemezdir. 

✅***Demet Tanımlama Yöntemleri***

```python
# 1. Yöntem : Parantez içinde tanımlama
gezegenler = ("Merkür", "Venüs", "Dünya", "Mars", "Jüpiter")   # Demetlerin elemanları parantez içinde yazılır ve virgül ile ayrılır.
```

####

```python
# 2. Yöntem : Parantezsiz tanımlama
denizler = "Akdeniz", "Ege", "Marmara", "Karadeniz"     # Parantez içinde yazılmayabilir de
```

####

```python
# 3. Yöntem : tupple() Fonksiyonu ile tanımlama
a = tuple('qwerty')          # tupple() fonksiyonuna bir string gönderilir
print ("Oluşan Demet = ",a)
```

####

```python
# Bu fonksiyonu kullanarak başka veri tiplerini demete dönüştürebilirsiniz:
sosyal = ["Instagram", "Twitter", "Facebook", "TikTok", "LinkedIn"]     # Bir liste oluşturduk
a = tuple(sosyal)                                                       # Listeyi demete çevirdik
print ("Oluşan Demet = ",a)
```

####



✅**Tek Öğeli bir Demet Tanımlamak:**

```python
# Python programlama dilinde tek öğeli bir demet oluşturma işlemi biraz ‘tuhaf’tır.
#Eğer tek öğeye sahip bir demet oluşturacaksak şöyle bir şey yazmalıyız:

a = ('ahmet',)     # elemanımızı yazdıktan sonra virgül koyuyoruz
#veya:
a = 'ahmet',
```

```python
a = ('ahmet')      # Bu şekilde tanımladığımız şey bir stringdir.

print(type(a))     # a değişkeninin türünü type() fonksiyonu ile kontrol edersek string olduğunu görürüz
```

####

✅**Demetlerin Öğelerine Erişmek:**

```python
# Stringlerde ve listelerde kulanılan yöntemin aynısını burada da kullanabiliriz.
# demet_adı [eleman_indeks_numarası]

satranç = ("Şah", "Vezir", "Kale", "Fil", "At", "Piyon")
print(satranç[0])       # Demetin ilk elemanını verir.      (Şah)
print(satranç[1])       # Demetin ikinci elemanına ulaşırız (Vezir)
print(satranç[2])       # Demetin üçüncü elemanına ulaşırız (Kale)
print(satranç[-1])       # Demetin son  elemanına ulaşırız (Piyon)
```

####

✅**Demetlerle Listelerin Birbirinden Farkı**:

**1. Değiştirilebilirlik (Mutability)**🔎

*Demet:*   Değiştirilemez (Immutable). Bir demet oluşturulduktan sonra içeriğine müdahale edemezsin. Eleman eklenemez, çıkarılamaz veya güncellenemez.

*Liste:*   Değiştirilebilirdir. Liste oluşturulduktan sonra eleman ekleyebilir (append), silebilir (remove) veya mevcut bir elemanı güncelleyebilirsin.



---


**2. Yazım Şekli (Sözdizimi)**🔎

*Demet:*    Normal parantez () ile tanımlanır.

*Liste:*    Köşeli parantez [] ile tanımlanır.



---




**3. Performans ve Hız**🔎

*Demet:*    Daha az bellek harcar ve çok daha hızlı çalışır. Eğer verilerin sabitse demet kullanmak performansı artırır.

*Liste:*    Daha fazla bellek (RAM) harcar ve işlemleri demetlere göre biraz daha yavaştır.


---





**4. Metot Çeşitliliği**🔎

*Demet:*    İçeriği değişmediği için sadece iki metodu vardır: count() ve index().

*Liste:*    İçeriği değiştirebildiği için çok sayıda metoda sahiptir (sort, reverse, clear, pop vb.).



####

✅**Demetlerin Metotları :**

Demetlerin tipinin bizi ilgilendiren iki metodu var:

📌 **1-  index() :** Bir demet öğesinin demet içindeki konumunu söyler.

```python
nota = ("Do", "Re", "Mi", "Fa", "Sol", "La", "Si")
print(nota.index("Mi"))       # Mi notasının bulunduğu sırayı verir (2)
```

####

📌 **2 count():**  Bir öğenin o veri tipi
içinde kaç kez geçtiğini söyler

```python
notalar = ("Do", "Re", "Do", "Re", "Mi","Mi","Mi","Re","Mi","Mi","Do","Do")
print("Do notası sayısı = ", notalar.count("Do"))
print ("Re notası sayısı = ", notalar.count("Re"))
print ("Mi notası sayısı = ", notalar.count("Mi"))
```
