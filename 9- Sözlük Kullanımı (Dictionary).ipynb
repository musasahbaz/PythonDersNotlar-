## 🟥 Python'da Sözlükler (Dictionary)

Python'da **sözlük (dictionary)** veri yapısı, verileri **anahtar (key) – değer (value)** 
çiftleri halinde saklamak için kullanılır.

Sözlükler:
- Süslü parantez `{ }` ile oluşturulur
- Her veri bir **anahtar : değer** şeklindedir
- Anahtarlar benzersizdir
- Değerler tekrar edebilir


```python
ogrenci = {
    "ad": "Ali",
    "soyad": "Yılmaz",
    "yas": 16,
    "sinif": "10A"
}

print(ogrenci)
```

#### 📘 Sözlükte bir değere erişmek için  --->  sozluk[anahtar]

```python
ogrenci = {"ad":"Ayşe","yas":15,"sehir":"İzmir"}

print(ogrenci["ad"])             # ad anahtarına ait değeri verir
print(ogrenci["yas"])            # anahtar yoksa bu yöntem hata verir. 
```

#### 📘 get() metodu ile güvenli erişim


```python
ogrenci = {"ad":"Mehmet","yas":17}

print(ogrenci.get("ad"))
print(ogrenci.get("okul"))   # Anahtar yoksa hata vermez
```

```python
ing_sözlük = {"dil": "language", "bilgisayar": "computer", "masa": "table"}
sorgu = input("Lütfen anlamını öğrenmek istediğiniz kelimeyi yazınız:")
print(ing_sözlük.get(sorgu, "Bu kelime veritabanımızda yoktur!"))            #Bulunamaması durumunda nasıl bir hata mesajı isteniyorsa yazılabilir
```

#### 📘 len() : sözlükteki eleman sayısını bulma

```python
urunler = {
    "elma":10,
    "armut":5,
    "muz":7
}

print(len(urunler))
```

### 📘 keys() : Anahtarları listeleme

```python
arabalar = {
    "marka":"Toyota",
    "model":"Corolla",
    "yil":2022
}

print(arabalar.keys())
```

### 📘 values() : Değerleri listeleme

```python
arabalar = {
    "marka":"Toyota",
    "model":"Corolla",
    "yil":2022
}

print(arabalar.values())
```

### 📘 items() : Anahtar ve değerleri birlikte alma

```python
ogrenci = {
    "ad":"Zeynep",
    "yas":16,
    "sehir":"Urla"
}

print(ogrenci.items())
```

### 📘 Sözlükte for döngüsü ile gezinme

```python
ogrenci = {
    "ad":"Can",
    "yas":17,
    "sehir":"Ankara"
}

for i in ogrenci:
    print(i, ":", ogrenci[i])
```

### 📘 items() ile anahtar ve değerleri birlikte yazdırma

```python
ogrenci = {
    "ad":"Ece",
    "yas":15,
    "sehir":"İzmir"
}

for k, v in ogrenci.items():
    print(k, ":", v)
```

### 📘 Sözlüğe yeni veri eklemek

```python
ogrenci = {"ad":"Ali","yas":16}

ogrenci["sehir"] = "İstanbul"       # [ ] içinde belirterek sözlüğe yeni eleman ve değeri eklenebilir

print(ogrenci)
```

### 📘 Değer güncelleme

```python
ogrenci = {"ad":"Ali","yas":16}

ogrenci["yas"] = 17

print(ogrenci)
```

### 📘 update() metodu

```python
ogrenci = {"ad":"Ayşe","yas":15}

ogrenci.update({"yas":16})
ogrenci.update({"sehir":"İzmir"})

print(ogrenci)
```

### 📘 pop() ile veri silme

```python
ogrenci = {"ad":"Ali","yas":16,"sehir":"Bursa"}

ogrenci.pop("yas")

print(ogrenci)
```

### 📘 popitem() son elemanı siler

```python
veri = {
    "a":1,
    "b":2,
    "c":3
}

veri.popitem()

print(veri)
```

### 📘 del komutu ile silme

```python
ogrenci = {"ad":"Ali","yas":16}

del ogrenci["yas"]                    # del ogrenci   komutu ile sözlüğün tamamı da silinebilir.

print(ogrenci)
```

### 📘 clear() sözlüğü tamamen temizler

```python
veri = {"a":1,"b":2,"c":3}

veri.clear()

print(veri)
```

### 📘 fromkeys()
Bu metot yeni bir sözlük oluştururken listeler veya demetlerden yararlanır.

```python
# 1. Temel Kullanım (Varsayılan Değer None) 

anahtarlar = ["ad", "soyad", "okul"]
kullanici = dict.fromkeys(anahtarlar)     #Eğer bir değer belirtmezsen, Python otomatik olarak tüm anahtarlara None değerini atar.

print(kullanici)
# Çıktı: {'ad': None, 'soyad': None, 'okul': None}
```

```python
#2. Sabit Bir Başlangıç Değeri Atama

oyuncular = ["Ali", "Ayşe", "Mehmet"]    # Başlangıç değeri belirtilirse her elemana o değer atanır
puanlar = dict.fromkeys(oyuncular, 0)

print(puanlar)
```

```python
#3. String (Metin) Kullanarak Sözlük Oluşturma

harfler = "ABC"
durum = dict.fromkeys(harfler, "Bekliyor")

print(durum)
# Çıktı: {'A': 'Bekliyor', 'B': 'Bekliyor', 'C': 'Bekliyor'}
```

```python
#4. Liste veya Tuple ile Kullanım
gunler = ("Pazartesi", "Salı", "Çarşamba")
program = dict.fromkeys(gunler, []) # Dikkat: Tüm anahtarlar aynı liste objesine bakar!

print(program)
# Çıktı: {'Pazartesi': [], 'Salı': [], 'Çarşamba': []}
```

```python
# 5. Mevcut Bir Sözlüğü Sıfırlama

eski_ayarlar = {"tema": "koyu", "dil": "tr", "bildirim": True}
yeni_ayarlar = dict.fromkeys(eski_ayarlar, "varsayilan")

print(yeni_ayarlar)
# Çıktı: {'tema': 'varsayilan', 'dil': 'varsayilan', 'bildirim': 'varsayilan'}
```

### 📘 setdefault()
Bu metot yardımıyla bir sözlük içinde arama yapabiliyor, eğer aradığımız anahtar sözlükte 
yoksa, setdefault() metodu içinde belirttiğimiz özellikleri taşıyan yeni bir anahtar-değer çifti
oluşturabiliyoruz.

```python
sepet = {"meyveler": ("elma", "armut"), "sebzeler": ("pırasa", "fasulye")}
sepet.setdefault("içecekler", ("su", "kola"))          # içecekler isminde bir değer olmadığı için oluşturulur ve sözlüğe eklenir

print(sepet)
```

### 📘 İç içe sözlükler (Nested Dictionary)

```python
ogrenciler = {
    "ogr1":{"ad":"Ali","yas":16},
    "ogr2":{"ad":"Ayşe","yas":15}
}

print(ogrenciler["ogr2"]["ad"])
```

```python
tel = {
    "iPhone 15": {"ekran": "6.1 inç", "islemci": "A16 Bionic", "depolama": [128, 256, 512]},
    "S24": {"ekran": "6.2 inç", "islemci": "Exynos 2400", "depolama": [128, 256]}
}

print(tel ["S24"]["islemci"])
print(tel ["iPhone 15"]["ekran"])
```

### 📘 copy ()  -Sözlük kopyalama 

```python
a = {"x":1,"y":2}

b = a.copy()

print(b)
```

### 📘 Sözlük Üreteçleri (Dictionary Comprehensions)

```python
# Alfabe harflerinin her birine bir sayı verelim.

harfler = 'abcçdefgğhıijklmnoöprsştuüvyz'

sira = {}
for i in range(len(harfler)):
    sira[harfler[i]] = i+1                # i 0'dan başlayacağı için 1 fazlası alındı

print( sira )
```

##### şimdi bu örneği sözlük üreteçleri ile daha kısa hale getirelim.

```python
harfler = 'abcçdefgğhıijklmnoöprsştuüvyz'
sira = {i: harfler.index(i) + 1 for i in harfler}
print(sira)
```

```python
# Verilen bir sözlükteki tüm elemanların karakter uzunluğunu bir sözlük haline getirelim

isimler = ["ahmet", "mehmet", "fırat", "zeynep", "selma", "abdullah", "cem"]
sözlük = {i: len(i) for i in isimler}
print(sözlük)
```
